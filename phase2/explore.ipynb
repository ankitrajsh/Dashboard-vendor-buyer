```python

def create_url_tree_from_merged_data():
    print(" Creating URL tree from merged_data...")
    
    # Use merged_data directly since it already contains both URL and visitor data
    session_urls = merged_data.dropna(subset=['full_url', 'idvisitor']).copy()
    
    print(f" Working with: {len(session_urls)} records with URL and visitor data")
    print(f" Total rows in merged_data: {len(merged_data)}")
    print(f" Rows with full_url: {merged_data['full_url'].notna().sum()}")
    
    if len(session_urls) == 0:
        print(" No records with both URL and visitor data found!")
        return {}
    
    # Create mapping from URL to visitors
    url_to_visitors = session_urls.groupby('full_url')['idvisitor'].apply(set).to_dict()
    print(f" Found {len(url_to_visitors)} unique URLs with visitor data")
    
    # Show sample URL-to-visitors mapping
    print("\n Sample URL-to-visitors mapping:")
    for i, (url, visitors) in enumerate(list(url_to_visitors.items())[:5]):
        print(f"  {url}: {len(visitors)} visitors")
    
    url_tree = {}
    
    for url, visitors in url_to_visitors.items():
        # Parse URL to get domain and path
        url_match = re.match(r'(https?://[^/]+)(/.*)?', str(url))
        
        if url_match:
            domain = url_match.group(1)
            path = url_match.group(2) if url_match.group(2) else '/'
            
            # Split path into segments
            if path == '/':
                segments = ['root']  # Root path
            else:
                segments = [seg for seg in path.split('/') if seg]
            
            # Build tree structure
            current_level = url_tree
            
            # First level: domain
            if domain not in current_level:
                current_level[domain] = {
                    '_visitors': set(),
                    '_urls': set(),
                    '_children': {}
                }
            current_level[domain]['_visitors'].update(visitors)
            current_level[domain]['_urls'].add(url)
            
            # Navigate to children for path segments
            current_level = current_level[domain]['_children']
            
            # Add path segments
            for segment in segments:
                if segment not in current_level:
                    current_level[segment] = {
                        '_visitors': set(),
                        '_urls': set(),
                        '_children': {}
                    }
                current_level[segment]['_visitors'].update(visitors)
                current_level[segment]['_urls'].add(url)
                current_level = current_level[segment]['_children']
    
    return url_tree
```

```python
# Function to analyze tree layers
def analyze_url_tree_layers(tree, max_level=3):
    layer_stats = {}
    
    def count_layer(node, current_level):
        if current_level > max_level:
            return
            
        if current_level not in layer_stats:
            layer_stats[current_level] = {
                'segments': [],
                'total_visitors': set(),
                'segment_details': {}
            }
        
        if isinstance(node, dict):
            for key, value in node.items():
                if not key.startswith('_'):  # Skip metadata keys
                    layer_stats[current_level]['segments'].append(key)
                    
                    if isinstance(value, dict) and '_visitors' in value:
                        visitors = value['_visitors']
                        layer_stats[current_level]['total_visitors'].update(visitors)
                        
                        # Store details for this segment
                        layer_stats[current_level]['segment_details'][key] = {
                            'visitors': len(visitors),
                            'urls': len(value.get('_urls', set()))
                        }
                    
                    # Recurse into children
                    if isinstance(value, dict) and '_children' in value:
                        count_layer(value['_children'], current_level + 1)
    
    count_layer(tree, 0)
    return layer_stats
```

```python
def create_detailed_url_tree_with_user_counts():
    
    # Use merged_data directly since it has both URL and visitor data
    session_urls = merged_data.dropna(subset=['full_url', 'idvisitor']).copy()
    
    print(f"Working with: {len(session_urls)} records")
    print(f"Unique visitors: {session_urls['idvisitor'].nunique()}")
    print(f"Unique URLs: {session_urls['full_url'].nunique()}")
    
    if len(session_urls) == 0:
        print(" No data available!")
        return {}
    
    # Create detailed tree structure
    url_tree = {}
    
    # Process each URL and track users at each level
    for _, row in session_urls.iterrows():
        url = row['full_url']
        visitor = row['idvisitor']
        
        # Parse URL to get domain and path
        url_match = re.match(r'(https?://[^/]+)(/.*)?', str(url))
        
        if url_match:
            domain = url_match.group(1)
            path = url_match.group(2) if url_match.group(2) else '/'
            
            # Split path into segments
            if path == '/':
                segments = ['root']
            else:
                segments = [seg for seg in path.split('/') if seg]
            
            # Initialize domain level
            if domain not in url_tree:
                url_tree[domain] = {
                    '_visitors': set(),
                    '_urls': set(),
                    '_pageviews': 0,
                    '_children': {}
                }
            
            # Add visitor to domain
            url_tree[domain]['_visitors'].add(visitor)
            url_tree[domain]['_urls'].add(url)
            url_tree[domain]['_pageviews'] += 1
            
            # Navigate through path segments
            current_level = url_tree[domain]['_children']
            
            for i, segment in enumerate(segments):
                if segment not in current_level:
                    current_level[segment] = {
                        '_visitors': set(),
                        '_urls': set(),
                        '_pageviews': 0,
                        '_children': {}
                    }
                
                # Add visitor to this segment
                current_level[segment]['_visitors'].add(visitor)
                current_level[segment]['_urls'].add(url)
                current_level[segment]['_pageviews'] += 1
                
                # Move to next level
                current_level = current_level[segment]['_children']
    
    return url_tree

# Create the detailed tree

```

```python
detailed_tree = create_detailed_url_tree_with_user_counts()
```

```python
# Function to analyze each layer with detailed metrics
def analyze_layers_with_user_counts(tree, max_level=4):
    layer_analysis = {}
    
    def traverse_tree(node, level, path=""):
        if level > max_level:
            return
            
        if level not in layer_analysis:
            layer_analysis[level] = {
                'segments': [],
                'total_unique_visitors': set(),
                'total_pageviews': 0,
                'segment_details': {}
            }
        
        if isinstance(node, dict):
            for key, value in node.items():
                if not key.startswith('_'):  # Skip metadata
                    segment_path = f"{path}/{key}" if path else key
                    
                    # Add to layer analysis
                    layer_analysis[level]['segments'].append(key)
                    
                    if isinstance(value, dict):
                        visitors = value.get('_visitors', set())
                        urls = value.get('_urls', set())
                        pageviews = value.get('_pageviews', 0)
                        
                        # Update layer totals
                        layer_analysis[level]['total_unique_visitors'].update(visitors)
                        layer_analysis[level]['total_pageviews'] += pageviews
                        
                        # Store segment details
                        layer_analysis[level]['segment_details'][key] = {
                            'unique_visitors': len(visitors),
                            'unique_urls': len(urls),
                            'total_pageviews': pageviews,
                            'path': segment_path
                        }
                        
                        # Recurse to children
                        if '_children' in value:
                            traverse_tree(value['_children'], level + 1, segment_path)
    
    traverse_tree(tree, 0)
    return layer_analysis
```

```python
#  SANKEY CHART IMPLEMENTATION - COMPLETE GUIDE
print(" SANKEY CHART IMPLEMENTATION FOR SUPERSET")

print(" STEP 1: COPY THIS SQL TO SUPERSET SQL LAB")

# Complete SQL query for Sankey chart
sankey_sql = """-- SANKEY CHART: USER FLOW ANALYSIS
-- This query creates source-target-value data for Sankey visualization

WITH page_categories AS (
    SELECT 
        lva.idvisit,
        lva.pageview_position,
        la.name as page_url,
        CASE 
            WHEN la.name ILIKE '%auth%' OR la.name ILIKE '%login%' OR la.name ILIKE '%signin%' THEN 'Authentication'
            WHEN la.name ILIKE '%dashboard%' OR la.name ILIKE '%management%' OR la.name ILIKE '%admin%' THEN 'Dashboard'
            WHEN la.name ILIKE '%product%' OR la.name ILIKE '%rfq%' OR la.name ILIKE '%catalog%' THEN 'Product/RFQ'
            WHEN la.name ILIKE '%auction%' OR la.name ILIKE '%deal%' OR la.name ILIKE '%bid%' THEN 'Auction/Deals'
            WHEN la.name ILIKE '%search%' OR la.name ILIKE '%browse%' THEN 'Search/Browse'
            WHEN la.name ILIKE '%profile%' OR la.name ILIKE '%settings%' OR la.name ILIKE '%account%' THEN 'Profile/Settings'
            WHEN la.name ILIKE '%news%' OR la.name ILIKE '%blog%' THEN 'News/Content'
            WHEN la.name ILIKE '%help%' OR la.name ILIKE '%support%' OR la.name ILIKE '%faq%' THEN 'Help/Support'
            WHEN la.name ILIKE '%home%' OR la.name = '/' OR la.name ILIKE '%index%' THEN 'Homepage'
            WHEN la.name ILIKE '%contact%' OR la.name ILIKE '%about%' THEN 'Company Info'
            ELSE 'Other Pages'
        END as page_category
    FROM matomo_log_link_visit_action lva
    LEFT JOIN matomo_log_action la ON lva.idaction_url = la.idaction
    WHERE la.name IS NOT NULL
        AND lva.server_time >= CURRENT_DATE - INTERVAL '60 days'  -- Last 60 days
),
user_flows AS (
    SELECT 
        p1.page_category as source,
        p2.page_category as target,
        p1.idvisit
    FROM page_categories p1
    JOIN page_categories p2 
        ON p1.idvisit = p2.idvisit 
        AND p2.pageview_position = p1.pageview_position + 1
    WHERE p1.page_category != p2.page_category  -- Only show transitions between different categories
        AND p1.page_category IS NOT NULL 
        AND p2.page_category IS NOT NULL
)
SELECT 
    source,
    target,
    COUNT(*) as value
FROM user_flows
GROUP BY source, target
HAVING COUNT(*) >= 10  -- Filter out flows with less than 10 transitions
ORDER BY COUNT(*) DESC
LIMIT 30;"""

print(sankey_sql)

print("\n STEP 2: SUPERSET CONFIGURATION")

config_guide = [
    "1. Chart Type: Select 'Sankey Diagram'",
    "2. Source Field: Choose 'source' column",
    "3. Target Field: Choose 'target' column", 
    "4. Metric: Select 'SUM(value)' or just 'value'",
    "5. Color Scheme: Choose your preferred colors"
]

for step in config_guide:
    print(f"   {step}")

print("\nSTEP 3: DETAILED IMPLEMENTATION PROCESS")

implementation_steps = [
    "1. Open Apache Superset in your browser",
    "2. Navigate to 'SQL Lab' → 'SQL Editor'",
    "3. Select your PostgreSQL database connection",
    "4. Copy and paste the SQL query above",
    "5. Click 'Run' to execute the query",
    "6. Verify the results show source, target, value columns",
    "7. Click 'Explore' button (chart icon) next to results",
    "8. In the chart creation page:",
    "   - Chart Type: Search for 'Sankey' or 'Flow'",
    "   - Source: Select 'source'",
    "   - Target: Select 'target'", 
    "   - Metric: Select 'SUM(value)'",
    "9. Click 'Run' to generate the visualization",
    "10. Customize colors, labels, and styling",
    "11. Save the chart with name 'User Flow - Sankey'"
]

for step in implementation_steps:
    print(f"   {step}")

print("\n WHAT YOUR SANKEY CHART WILL REVEAL:")

insights = [
    "• Most popular user journey paths through your app",
    "• Where users typically enter your application",
    "• Common page-to-page transition patterns",
    "• Identification of drop-off and exit points",
    "• Feature usage and navigation patterns",
    "• User flow bottlenecks and optimization opportunities"
]

for insight in insights:
    print(f"   {insight}")

print("\n TROUBLESHOOTING TIPS:")

troubleshooting = [
    "• If Sankey isn't available: Try 'Chord Diagram' or 'Directed Graph'",
    "• If query is slow: Reduce date range or add more filters",
    "• If too many flows: Increase HAVING COUNT(*) >= 10 to higher number",
    "• If chart is cluttered: Reduce LIMIT 30 to LIMIT 15",
    "• For better categories: Adjust the CASE WHEN conditions in SQL"
]

for tip in troubleshooting:
    print(f"   {tip}")

print("\n EXPECTED DATA SAMPLE:")
print("source          | target          | value")
print("--------------- | --------------- | -----")
print("Homepage        | Search/Browse   | 245")
print("Authentication  | Dashboard       | 189") 
print("Search/Browse   | Product/RFQ     | 167")
print("Dashboard       | Product/RFQ     | 143")
print("Product/RFQ     | Authentication  | 98")

print("\n STYLING RECOMMENDATIONS:")

styling_tips = [
    "• Use distinct colors for different page categories",
    "• Set node width based on traffic volume", 
    "• Enable labels for all nodes",
    "• Use curved links for better readability",
    "• Add title: 'User Journey Flow Analysis'"
]

for tip in styling_tips:
    print(f"   {tip}")

print("\n READY TO CREATE YOUR SANKEY CHART!")
print(" Copy the SQL above and follow the steps in Superset!")
```
