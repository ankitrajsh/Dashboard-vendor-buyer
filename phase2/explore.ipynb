```python
# Assign shorter variable names for easier analysis
logs = matomo_log_visit
link = matomo_log_link_visit_action
action = matomo_log_action

print("Data assigned:")
print(f"logs: {logs.shape}")
print(f"link: {link.shape}")
print(f"action: {action.shape}")
```

```python
# Create full URLs from action table
url_prefix_map = {
    0: 'http://',
    1: 'http://www.',
    2: 'https://',
    3: 'https://www.'
}

action['url_prefix_clean'] = action['url_prefix'].fillna(0).astype(int)
action['url_prefix_text'] = action['url_prefix_clean'].map(url_prefix_map)
action['full_url'] = action['url_prefix_text'] + action['name']

print(f"Created {action['full_url'].nunique()} unique URLs")
print("Sample URLs:")
print(action['full_url'].head(10).tolist())
```

```python
# Create session analysis with URL information
merged_data = link.merge(action[['idaction', 'full_url']], 
                        left_on='idaction_url', 
                        right_on='idaction', 
                        how='left')

print(f"Merged data shape: {merged_data.shape}")
print(f"Sessions with URLs: {merged_data['full_url'].notna().sum()}")

# Sort by session and pageview position to get user journey
merged_data = merged_data.sort_values(['idvisit', 'pageview_position'])

print("Sample merged data:")
print(merged_data[['idvisit', 'pageview_position', 'full_url', 'time_spent_ref_action']].head())
```

```python
# Analyze user journeys within sessions
session_journeys = merged_data.groupby('idvisit').agg({
    'full_url': lambda x: list(x.dropna()),
    'pageview_position': 'max',
    'time_spent_ref_action': 'sum',
    'server_time': ['min', 'max']
}).reset_index()

session_journeys.columns = ['idvisit', 'journey_urls', 'total_pages', 'total_time_spent', 'session_start', 'session_end']

print(f"Analyzed {len(session_journeys)} sessions")
print(f"Average pages per session: {session_journeys['total_pages'].mean():.2f}")
print(f"Max pages in a session: {session_journeys['total_pages'].max()}")

print("\nSample session journeys:")
for i in range(5):
    session = session_journeys.iloc[i]
    print(f"Session {session['idvisit']}: {len(session['journey_urls'])} pages")
    for j, url in enumerate(session['journey_urls'][:3], 1):
        print(f"  {j}. {url}")
    if len(session['journey_urls']) > 3:
        print(f"  ... and {len(session['journey_urls']) - 3} more")
```

```python
# Calculate page importance based on visit frequency
page_visits = merged_data['full_url'].value_counts()
total_pageviews = len(merged_data)

page_importance = pd.DataFrame({
    'url': page_visits.index,
    'visit_count': page_visits.values,
    'visit_percentage': (page_visits.values / total_pageviews) * 100
})

print("Top 10 Most Visited Pages:")
print(page_importance.head(10))

print(f"\nTotal unique pages: {len(page_importance)}")
print(f"Pages visited by >1% of pageviews: {(page_importance['visit_percentage'] > 1).sum()}")
print(f"Pages visited by >5% of pageviews: {(page_importance['visit_percentage'] > 5).sum()}")
```

```python
# Analyze entry and exit pages importance
entry_pages = merged_data[merged_data['pageview_position'] == 1]['full_url'].value_counts()
exit_pages = merged_data.groupby('idvisit')['full_url'].last().value_counts()

print("Top 10 Entry Pages:")
for i, (url, count) in enumerate(entry_pages.head(10).items(), 1):
    print(f"{i:2d}. {url}: {count} sessions")

print("\nTop 10 Exit Pages:")
for i, (url, count) in enumerate(exit_pages.head(10).items(), 1):
    print(f"{i:2d}. {url}: {count} sessions")

# Calculate bounce rate for entry pages
session_lengths = merged_data.groupby('idvisit')['pageview_position'].max()
single_page_sessions = (session_lengths == 1).sum()
total_sessions = len(session_lengths)

print(f"\nBounce Rate Analysis:")
print(f"Single page sessions: {single_page_sessions}")
print(f"Total sessions: {total_sessions}")
print(f"Overall bounce rate: {(single_page_sessions/total_sessions)*100:.2f}%")
```

```python
# Analyze page transition patterns
transitions = []

for session in session_journeys['journey_urls']:
    if len(session) > 1:
        for i in range(len(session) - 1):
            from_page = session[i]
            to_page = session[i + 1]
            transitions.append((from_page, to_page))

transition_df = pd.DataFrame(transitions, columns=['from_page', 'to_page'])
transition_counts = transition_df.groupby(['from_page', 'to_page']).size().reset_index(name='count')
transition_counts = transition_counts.sort_values('count', ascending=False)

print("Top 10 Page Transitions:")
for i, row in transition_counts.head(10).iterrows():
    print(f"{row['count']:3d}. {row['from_page']} -> {row['to_page']}")

print(f"\nTotal unique transitions: {len(transition_counts)}")
print(f"Most common transition occurs {transition_counts['count'].max()} times")
```

```python
# Calculate time spent analysis for page importance
time_analysis = merged_data.groupby('full_url').agg({
    'time_spent_ref_action': ['count', 'sum', 'mean', 'median']
}).reset_index()

time_analysis.columns = ['url', 'visit_count', 'total_time', 'avg_time', 'median_time']
time_analysis = time_analysis.sort_values('total_time', ascending=False)

print("Top 10 Pages by Total Time Spent:")
for i, row in time_analysis.head(10).iterrows():
    print(f"{row['total_time']:8.0f}s total, {row['avg_time']:6.1f}s avg: {row['url']}")

print("\nTop 10 Pages by Average Time Spent:")
time_analysis_avg = time_analysis.sort_values('avg_time', ascending=False)
for i, row in time_analysis_avg.head(10).iterrows():
    print(f"{row['avg_time']:6.1f}s avg ({row['visit_count']} visits): {row['url']}")

# Filter pages with reasonable visit counts
popular_pages = time_analysis[time_analysis['visit_count'] >= 10]
print(f"\nPages with 10+ visits: {len(popular_pages)}")
print("Top engagement pages (10+ visits, by avg time):")
for i, row in popular_pages.sort_values('avg_time', ascending=False).head(5).iterrows():
    print(f"{row['avg_time']:6.1f}s avg ({row['visit_count']} visits): {row['url']}")
```

```python
# Analyze page depth and position importance
position_analysis = merged_data.groupby('full_url')['pageview_position'].agg(['mean', 'min', 'max', 'count']).reset_index()
position_analysis.columns = ['url', 'avg_position', 'first_position', 'last_position', 'visit_count']

print("Pages typically visited early in sessions (low avg position):")
early_pages = position_analysis.sort_values('avg_position').head(10)
for i, row in early_pages.iterrows():
    print(f"Pos {row['avg_position']:.1f} ({row['visit_count']} visits): {row['url']}")

print("\nPages typically visited later in sessions (high avg position):")
late_pages = position_analysis.sort_values('avg_position', ascending=False).head(10)
for i, row in late_pages.iterrows():
    print(f"Pos {row['avg_position']:.1f} ({row['visit_count']} visits): {row['url']}")

# Pages that appear in different positions (versatile pages)
position_analysis['position_range'] = position_analysis['last_position'] - position_analysis['first_position']
versatile_pages = position_analysis[position_analysis['visit_count'] >= 10].sort_values('position_range', ascending=False)

print("\nMost versatile pages (visited at various session depths):")
for i, row in versatile_pages.head(5).iterrows():
    print(f"Range {row['position_range']:.0f} ({row['visit_count']} visits): {row['url']}")
```

```python

page_visits = merged_data['full_url'].value_counts()
total_pageviews = len(merged_data)

page_importance = pd.DataFrame({
    'url': page_visits.index,
    'visit_count': page_visits.values,
    'visit_percentage': (page_visits.values / total_pageviews) * 100
})

page_scores = page_importance.copy()

# Add time metrics
page_scores = page_scores.merge(time_analysis[['url', 'total_time', 'avg_time']], on='url', how='left')

# Add position metrics
page_scores = page_scores.merge(position_analysis[['url', 'avg_position']], on='url', how='left')

# Add entry/exit page counts
entry_counts = entry_pages.reset_index()
entry_counts.columns = ['url', 'entry_count']
exit_counts = exit_pages.reset_index()
exit_counts.columns = ['url', 'exit_count']

page_scores = page_scores.merge(entry_counts, on='url', how='left')
page_scores = page_scores.merge(exit_counts, on='url', how='left')
page_scores[['entry_count', 'exit_count']] = page_scores[['entry_count', 'exit_count']].fillna(0)

# Calculate composite importance score
# Normalize metrics to 0-1 scale
page_scores['visit_score'] = page_scores['visit_count'] / page_scores['visit_count'].max()
page_scores['time_score'] = page_scores['total_time'] / page_scores['total_time'].max()
page_scores['entry_score'] = page_scores['entry_count'] / page_scores['entry_count'].max()

# Combined importance score (weighted average)
page_scores['importance_score'] = (
    page_scores['visit_score'] * 0.4 +
    page_scores['time_score'] * 0.3 +
    page_scores['entry_score'] * 0.3
)

page_scores = page_scores.sort_values('importance_score', ascending=False)

print("Top 15 Most Important Pages (Composite Score):")
for i, row in page_scores.head(15).iterrows():
    print(f"{row['importance_score']:.3f}: {row['url']} ({row['visit_count']} visits, {row['total_time']:.0f}s total)")
```
