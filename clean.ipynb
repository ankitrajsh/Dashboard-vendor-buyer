```python
# Convert server_time to datetime for proper analysis
link['server_time_360'] = pd.to_datetime(link['server_time'])

# Analyze all time-related columns in link dataframe
print("Time-related columns analysis in link dataframe:")
print("="*60)

# 1. Server time analysis
print("\n1. SERVER_TIME Analysis:")
print(f"   Data type: {link['server_time'].dtype}")
print(f"   Min date: {link['server_time'].min()}")
print(f"   Max date: {link['server_time'].max()}")
print(f"   Date range: {(pd.to_datetime(link['server_time'].max()) - pd.to_datetime(link['server_time'].min())).days} days")
print(f"   Unique timestamps: {link['server_time'].nunique()}")

# 2. Time performance columns analysis
time_columns = ['time_dom_completion', 'time_dom_processing', 'time_network', 
                'time_on_load', 'time_server', 'time_transfer']

print("\n2. TIME PERFORMANCE METRICS:")
for col in time_columns:
    if col in link.columns:
        print(f"\n   {col.upper()}:")
        print(f"     Non-null count: {link[col].count()}")
        if link[col].count() > 0:
            print(f"     Mean: {link[col].mean():.2f} ms")
            print(f"     Median: {link[col].median():.2f} ms")
            print(f"     Min: {link[col].min():.2f} ms")
            print(f"     Max: {link[col].max():.2f} ms")
            print(f"     Std: {link[col].std():.2f} ms")
        else:
            print(f"     No data available")

# 3. Time spent analysis
print("\n3. TIME_SPENT_REF_ACTION Analysis:")
print(f"   Non-null count: {link['time_spent_ref_action'].count()}")
if link['time_spent_ref_action'].count() > 0:
    print(f"   Mean time spent: {link['time_spent_ref_action'].mean():.2f} seconds")
    print(f"   Median time spent: {link['time_spent_ref_action'].median():.2f} seconds")
    print(f"   Min time spent: {link['time_spent_ref_action'].min()} seconds")
    print(f"   Max time spent: {link['time_spent_ref_action'].max()} seconds")
    print(f"   Std time spent: {link['time_spent_ref_action'].std():.2f} seconds")

# 4. Hourly distribution of server_time
print("\n4. HOURLY DISTRIBUTION:")
link['hour'] = pd.to_datetime(link['server_time']).dt.hour
hourly_dist = link['hour'].value_counts().sort_index()
print("   Hour | Count")
print("   -----|------")
for hour, count in hourly_dist.items():
    print(f"   {hour:2d}   | {count}")

# 5. Daily distribution 
print("\n5. DAILY DISTRIBUTION:")
daily_dist = link.groupby(link['server_time_360'].dt.date).size()
print("   Date       | Count")
print("   -----------|------")
for date, count in daily_dist.items():
    print(f"   {date} | {count}")

# 6. Performance metrics correlation
print("\n6. PERFORMANCE METRICS SUMMARY:")
perf_cols = [col for col in time_columns if col in link.columns and link[col].count() > 0]
if perf_cols:
    perf_data = link[perf_cols].describe()
    print(perf_data)
```

```python
# Merge the dataframes to create a comprehensive dataset for session analysis
merged_data = logs.merge(link, on='idvisit', how='inner')

# Convert server_time to datetime if not already done
merged_data['server_time_dt'] = pd.to_datetime(merged_data['server_time'])

# Create session time analysis
session_analysis = merged_data.groupby('idvisit').agg({
    'server_time_dt': ['min', 'max', 'count'],
    'time_spent_ref_action': 'sum',
    'visit_total_time': 'first',
    'pageview_position': 'max'
}).reset_index()

# Flatten column names
session_analysis.columns = ['idvisit', 'session_start', 'session_end', 'total_pageviews', 
                           'total_time_spent_on_pages', 'visit_total_time', 'max_pageview_position']

# Calculate session duration in seconds
session_analysis['session_duration_seconds'] = (
    session_analysis['session_end'] - session_analysis['session_start']
).dt.total_seconds()

# Calculate session duration in minutes
session_analysis['session_duration_minutes'] = session_analysis['session_duration_seconds'] / 60

print("Session Time Analysis:")
print("=" * 50)
print(f"Total sessions: {len(session_analysis)}")
print(f"Average session duration: {session_analysis['session_duration_minutes'].mean():.2f} minutes")
print(f"Median session duration: {session_analysis['session_duration_minutes'].median():.2f} minutes")
print(f"Max session duration: {session_analysis['session_duration_minutes'].max():.2f} minutes")
print(f"Min session duration: {session_analysis['session_duration_minutes'].min():.2f} minutes")

print("\nSession Duration Distribution:")
print(session_analysis['session_duration_minutes'].describe())

# Display first few sessions
print("\nFirst 10 sessions:")
print(session_analysis.head(10))

# Add custom_dimension_1 to session analysis
session_analysis = session_analysis.merge(
    logs[['idvisit', 'custom_dimension_1']], 
    on='idvisit', 
    how='left'
)

print("\nCustom Dimension 1 Analysis:")
print("=" * 50)
print(f"Sessions with custom_dimension_1: {session_analysis['custom_dimension_1'].notna().sum()}")
print(f"Sessions without custom_dimension_1: {session_analysis['custom_dimension_1'].isna().sum()}")
print("\nCustom Dimension 1 Distribution:")
print(session_analysis['custom_dimension_1'].value_counts(dropna=False))

# Add URL analysis to session analysis by merging with action data
session_urls = merged_data.merge(
    action[['idaction_url', 'full_url']], 
    left_on='idaction_url', 
    right_on='idaction_url', 
    how='left'
)

# Group URLs by session
session_url_analysis = session_urls.groupby('idvisit').agg({
    'full_url': lambda x: list(x.dropna().unique()),
    'pageview_position': 'count'
}).reset_index()

session_url_analysis.columns = ['idvisit', 'urls_visited', 'total_url_visits']

# Merge with existing session analysis
session_analysis = session_analysis.merge(session_url_analysis, on='idvisit', how='left')

print("\nURL Analysis per Session:")
print("=" * 50)
print(f"Sessions with URL data: {session_analysis['urls_visited'].notna().sum()}")
print(f"Average URLs per session: {session_analysis['total_url_visits'].mean():.2f}")
print(f"Max URLs in a session: {session_analysis['total_url_visits'].max()}")

print("\nSample sessions with URLs:")
for i, row in session_analysis.head(10).iterrows():
    print(f"\nSession {row['idvisit']} (Duration: {row['session_duration_minutes']:.1f} min):")
    if pd.notna(row['urls_visited']) and row['urls_visited']:
        for j, url in enumerate(row['urls_visited'][:5], 1):  # Show first 5 URLs
            print(f"  {j}. {url}")
        if len(row['urls_visited']) > 5:
            print(f"  ... and {len(row['urls_visited']) - 5} more URLs")
    else:
        print("  No URL data available")
```

```python
import re

# Create a tree structure showing URL paths after each slash count

def create_url_tree():
    # Filter out NaN values and extract the path after domain
    valid_urls = action.dropna(subset=['full_url'])
    
    url_tree = {}
    
    for url in valid_urls['full_url']:
        # Remove protocol and domain to get the path
        path_match = re.match(r'https?://[^/]+(/.*)?', url)
        if path_match and path_match.group(1):
            path = path_match.group(1)
            # Split by slash and remove empty strings
            segments = [seg for seg in path.split('/') if seg]
            
            # Build tree structure
            current_level = url_tree
            for i, segment in enumerate(segments):
                if segment not in current_level:
                    current_level[segment] = {}
                current_level = current_level[segment]
    
    return url_tree

# Function to print tree structure
def print_tree(tree, level=0, max_level=3):
    if level > max_level:
        return
    # Count segments at current level before last slash
    segment_count = len(tree)
    if level == 0:
        print(f"Total URL segments: {segment_count}")
    segment_count = len(tree)
    if level == 0:
        print(f"Total URL segments: {segment_count}")
    for key, subtree in sorted(tree.items()):
        print("  " * level + f"├── {key}")
        if isinstance(subtree, dict) and subtree:
            print_tree(subtree, level + 1, max_level)
# Create and display the URL tree
url_tree = create_url_tree()
print("URL Path Tree Structure:")
print_tree(url_tree)

# Also show slash count distribution
print("\n" + "="*50)
print("Slash Count Distribution:")
print(action['slash_count'].value_counts().sort_index())
```

```python
#how many unique in first level
def count_unique_first_level(tree):
    if not tree:
        return 0
    # Count unique keys at the first level
    return len(tree)

unique_count = count_unique_first_level(url_tree)
print(f"Unique URL segments at first level: {unique_count}")
#also print the all unique first level segments
print("Unique URL segments at first level:")
for key in sorted(url_tree.keys()):
    print(f"├── {key}") 
```

```python
# Count unique segments at second level for each first level segment
def count_unique_second_level(tree):
    second_level_counts = {}
    total_unique_second_level = set()
    
    for first_key, first_subtree in tree.items():
        if isinstance(first_subtree, dict):
            second_level_count = len(first_subtree)
            second_level_counts[first_key] = second_level_count
            # Add all second level keys to the set for total unique count
            total_unique_second_level.update(first_subtree.keys())
    
    return second_level_counts, len(total_unique_second_level)

second_level_counts, total_unique_second = count_unique_second_level(url_tree)

print("Unique URL segments at second level for each first level:")
for first_key, count in sorted(second_level_counts.items()):
    print(f"├── {first_key}: {count} unique segments")
    
print(f"\nTotal unique segments at second level across all first level segments: {total_unique_second}")

# print("\nDetailed second level segments:")
# for first_key, first_subtree in sorted(url_tree.items()):
#     if isinstance(first_subtree, dict) and first_subtree:
#         print(f"\n├── {first_key}")
#         for second_key in sorted(first_subtree.keys()):
#             print(f"    ├── {second_key}")
```

```python
# Count unique segments at third level for each second level segment
def count_unique_third_level(tree):
    third_level_counts = {}
    total_unique_third_level = set()
    
    for first_key, first_subtree in tree.items():
        if isinstance(first_subtree, dict):
            third_level_counts[first_key] = {}
            for second_key, second_subtree in first_subtree.items():
                if isinstance(second_subtree, dict):
                    third_level_count = len(second_subtree)
                    third_level_counts[first_key][second_key] = third_level_count
                    # Add all third level keys to the set for total unique count
                    total_unique_third_level.update(second_subtree.keys())
    
    return third_level_counts, len(total_unique_third_level)

third_level_counts, total_unique_third = count_unique_third_level(url_tree)

print("Unique URL segments at third level for each second level:")
for first_key, second_dict in sorted(third_level_counts.items()):
    print(f"\n├── {first_key}")
    for second_key, count in sorted(second_dict.items()):
        print(f"    ├── {second_key}: {count} unique segments")
    
print(f"\nTotal unique segments at third level across all segments: {total_unique_third}")
```

```python
# Count unique segments at fourth level for each third level segment
def count_unique_fourth_level(tree):
    fourth_level_counts = {}
    total_unique_fourth_level = set()
    
    for first_key, first_subtree in tree.items():
        if isinstance(first_subtree, dict):
            fourth_level_counts[first_key] = {}
            for second_key, second_subtree in first_subtree.items():
                if isinstance(second_subtree, dict):
                    fourth_level_counts[first_key][second_key] = {}
                    for third_key, third_subtree in second_subtree.items():
                        if isinstance(third_subtree, dict):
                            fourth_level_count = len(third_subtree)
                            fourth_level_counts[first_key][second_key][third_key] = fourth_level_count
                            # Add all fourth level keys to the set for total unique count
                            total_unique_fourth_level.update(third_subtree.keys())
    
    return fourth_level_counts, len(total_unique_fourth_level)

fourth_level_counts, total_unique_fourth = count_unique_fourth_level(url_tree)

# Only print segments that have higher number of fourth level segments (>5)
print("Unique URL segments at fourth level (showing only segments with >5 unique fourth level segments):")
for first_key, second_dict in sorted(fourth_level_counts.items()):
    first_level_printed = False
    for second_key, third_dict in sorted(second_dict.items()):
        second_level_printed = False
        for third_key, count in sorted(third_dict.items()):
            if count > 5:
                if not first_level_printed:
                    print(f"\n├── {first_key}")
                    first_level_printed = True
                if not second_level_printed:
                    print(f"    ├── {second_key}")
                    second_level_printed = True
                print(f"        ├── {third_key}: {count} unique segments")
    
print(f"\nTotal unique segments at fourth level across all segments: {total_unique_fourth}")
```
